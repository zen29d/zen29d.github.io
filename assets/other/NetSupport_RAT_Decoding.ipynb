## Reading NetSupport RAT sample

```python
with open('netsupport_loader.vbs') as file:
    sample = file.read()
```

```python
!shasum -a 256 netsupport_loader.vbs
```

```python
print(f'{sample}')
```

## Finding payloads using regex

```python
import re
```

```python
sub_key = re.search(r'\w+ = (\d+)', sample).group(1)
sub_key
```

```python
stage1_payload = re.findall(r'\((\d+[,\d+]*)\)', sample)
stage1_payload
```

```python
stage1_payload = [list(eval(line)) for line in stage1_payload] # string to list
```

## Initial payload

```python
stage1_decoded = ""
for code in stage1_payload:
    temp_code = "".join([chr(byt-int(sub_key)) for byt in code])
    stage1_decoded = stage1_decoded + temp_code + "\n"
```

```python
print(f'{stage1_decoded}')
```

## Stage 2 payload

```python
stage2_payload = re.findall(r'(\'[a-zA-Z0-9+/=]+\')', stage1_decoded) # extract base64 code
```

```python
base64_payload = eval(stage2_payload[0]) # first base64 code is payload
base64_payload
```

```python
aes_key = eval(stage2_payload[1]) # second base64 code is decryption key
aes_key
```

<p>Decoding base64 payload, which will result encypted payload</p>

```python
from base64 import b64decode 
from hexdump import hexdump
import filetype
```

```python
encry_payload = b64decode(base64_payload)
```

## Decrypting the payload

<p>First 16 bytes are initilizing vector in AES, however we dont need init vector for ECB mode</p>

```python
encry_payload = encry_payload[16:] # drop first 16 bytes from payload
```

```python
from Crypto.Cipher import AES
```

```python
cipher = AES.new(b64decode(aes_key), AES.MODE_ECB) # set the AES Cipher mode
```

```python
decry_payload = cipher.decrypt(encry_payload)
hexdump(decry_payload[:64])
```

```python
filetype.guess(decry_payload) # file is compressed
```

## Decompressing the payload

```python
import gzip
```

```python
decom_payload = gzip.decompress(decry_payload)
```

```python
decom_payload = (str(decom_payload))
```

```python
print(f"{decom_payload}")
```

## Extracting stage 3 payload

```python
stage3_payload = re.findall(r'\((\d+[,\d+]*)\)', decom_payload)
stage3_payload = [list(eval(code)) for code in stage3_payload]
```

## Decoding the stage 3 payload

```python
ioc = ""
for code in stage3_payload:
    temp_code = "".join([chr(byt-4274) for byt in code])
    ioc = ioc + temp_code + "\n"
```

```python
print(f'{ioc}')
```
